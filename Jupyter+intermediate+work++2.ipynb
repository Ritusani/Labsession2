```python
import pandas as pd
import csv
```

```python
df = pd.read_excel("data_breaches.xlsx")
```

```python
del df['alternative name']
```

```python
df  = df[['Entity','YEAR','records lost','ORGANISATION','METHOD OF LEAK','DATA SENSITIVITY']]
```

```python
df = df.rename(columns={'story':'Story',
                        'YEAR':'Year',
                        'records lost':'Records Lost',
                        'ORGANISATION':'Organisation',
                        'METHOD OF LEAK':'Method Of Leak',
                        'DATA SENSITIVITY':'Data Sensitivity',
                        })
```

```python
df = df.dropna(axis=1, how='all')
```

```python
updated_year = {"Year":{0:2004,1:2005,2:2006,3:2007,4:2008,5:2009,6:2010,7:2011,8:2012,9:2013,10:2014,11:2015,12:2016,13:2017,14:'Latest'}}
```

```python
df = df.replace(updated_year)
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.drop(df.index[0])
```

```python
data_sensitivity_updated = {"Data Sensitivity":{1:'Email Address/Online info',20:'SSN/Personal Details',300:'Credit card info',4000:'Passwords/Health Records',50000:'Bank Account Details'}}
```

```python
df = df.replace(data_sensitivity_updated)
```
